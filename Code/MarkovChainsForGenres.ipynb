```python
# input: structure_dict,
def create_markov_chain(my_dict):
    # convert dictionary to pandas dataframe
    dict_df = pd.DataFrame.from_dict(my_dict).fillna(0)

    # convert to proportions
    for col in dict_df.columns:
        dict_df[col] = dict_df[col]/sum(dict_df[col])

    # transpose so it matches standard format of markov chain
    dict_df = dict_df.transpose()
    return(dict_df)
```

```python
# return values if the EXACT genre is included:
def contains_genre(df, genre):
    return(df[[True if x >=1 else False for x in np.sum(df.values == genre, 1)]])

# return values where the exact genre or something similar is included:
def contains_genre_lax(df, genre):
    return(df[[True if x >=1 else False for x in np.sum([[genre in y if isinstance(y, str) else False for y in x[range(1,6)]] for x in genre_info.values],1)]])
```

```python
contains_genre(genre_info, "glam rock")
```

```python
folk_music= contains_genre_lax(genre_info, "folk")
```

```python
folk_files = ['/Volumes/SECONDDRIVE/prog/ug/chord_dicts/' + str(folk_music.iloc[i]['Decade']) + "/" + folk_music.iloc[i]['index'] + ".json" for i in range(len(folk_music))]
```

```python
folk_files = [x for x in folk_files if os.path.isfile(x)]
```

```python
song_structure = mf.clean_structure_dict(mf.get_structure_dict(folk_files), 10)
```

```python
create_markov_chain(song_structure)
```

```python
for genre in ["rock", "pop", "folk", "dance", "disco", "funk", "electronic", "hip-hop", 
              "rnb", "alternative", "indie", "grunge", "singer-songwriter", "new wave", "classic rock"]:
    genre_music = contains_genre(genre_info, genre)
    genre_files = ['/Volumes/SECONDDRIVE/prog/ug/chord_dicts/' + str(genre_music.iloc[i]['Decade']) + "/" + genre_music.iloc[i]['index'] + ".json" for i in range(len(genre_music))]
    genre_files = [x for x in genre_files if os.path.isfile(x)]
    song_structure = mf.clean_structure_dict(mf.get_structure_dict(genre_files), 10)
    song_chain = create_markov_chain(song_structure)
    song_chain.to_csv("../Output/SongStructure/genre_dist/" + genre + ".csv")
```
